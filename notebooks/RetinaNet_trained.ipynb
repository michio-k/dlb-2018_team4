# Multiple ingredients recognition

```python
import os
from PIL import Image
from IPython.display import display
from imageai.Detection import ObjectDetection
```

```python
model_dir = '/root/userspace/models/'
detector = ObjectDetection()
detector.setModelTypeAsRetinaNet()
detector.setModelPath( os.path.join( model_dir, "resnet50_coco_best_v2.0.1.h5"))
detector.loadModel()

images_dir = '/root/userspace/dlb-2018_team4/images/input/'
```

## example1

```python
output_image_path=os.path.join(images_dir , "detected_image1.jpg")
detections = detector.detectObjectsFromImage(input_image=os.path.join(images_dir , "meat_vegetables.jpg"), output_image_path=output_image_path)
display(Image.open(output_image_path))
```

```python
for eachObject in detections:
    print(eachObject["name"] , " : " , eachObject["percentage_probability"] )
```
