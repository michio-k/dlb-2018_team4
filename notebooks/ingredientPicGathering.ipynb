```python
from icrawler.builtin import GoogleImageCrawler

foods = ['じゃがいも', 'りんご', 'にんじん', 'たまねぎ', 'キャベツ', '白菜', '青菜', 'しらたき', 'ネギ']

for food in foods:
    crawler = GoogleImageCrawler(storage={"root_dir": food})
    crawler.crawl(keyword=food, max_num=200)
```
